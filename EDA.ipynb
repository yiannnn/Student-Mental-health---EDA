```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data=pd.read_csv('Student Mental health.csv')
data.head()
```

```python
data.info()
```

```python
top10_category_counts = data['What is your course?'].value_counts()

top10_categories = top10_category_counts.head(10)

other_categories = top10_category_counts.tail(top10_category_counts.size - 10).sum()
top10_categories['Other'] = other_categories

plt.figure(figsize=(12, 6))
sns.barplot(x=top10_categories.index, y=top10_categories.values, palette='flare')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Course', fontsize=12)
plt.title('What is your course?')
plt.show()
```

```python
numerical_cols = data.select_dtypes(include=['float64', 'int64'])

num_cols = numerical_cols.columns

for i, col in enumerate(num_cols):
    plt.figure(figsize=(12, 6))
    sns.histplot(numerical_cols[col], bins=7)
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df = data.copy()
df.drop('What is your course?', axis=1, inplace=True)

non_numerical_cols = df.select_dtypes(exclude=['float64', 'int64'])

cat_cols = non_numerical_cols.columns

for i, col in enumerate(cat_cols):
    plt.figure(figsize=(12, 6))
    non_num_categories = len(data[col].unique())
    sns.countplot(x=non_numerical_cols[col].values,data=df, palette="flare")
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='-', alpha=0.7)
    plt.title(f'{col}', fontsize=14)
plt.show()
```
